```python
#########################################################################################################
#################### Now we group the weeks by the day of the week for each participant #################
#########################################################################################################
df_participants = new_dataset.ID_DIAMANTE.unique() # 76 participants with different id
all_weekly_steps = []
all_weekly_goals = []
all_weekly_counts = []
all_weekly_average_steps = []
all_weekly_average_goals = []

for participant in df_participants:
    day_of_week_participant = new_dataset[new_dataset.ID_DIAMANTE == participant].day_of_the_week.reset_index(drop=True)
    today_steps_participant = new_dataset[new_dataset.ID_DIAMANTE == participant].today_steps.reset_index(drop=True)
    daily_goal_participant = new_dataset[new_dataset.ID_DIAMANTE == participant].daily_goal.reset_index(drop=True)

    
    weekly_steps = np.zeros(8)   # 8 is just a arbitruary number first, can change later on
    weekly_goals = np.zeros(8)
    weekly_counts = np.zeros(8) 
    weekly_average_steps = np.zeros(8)
    weekly_average_goals = np.zeros(8)
    
    week = 1   # We start off with week 1 
    count = 0
    for day in range(len(day_of_week_participant)):

        if day_of_week_participant[day] == 0:
            weekly_steps[week-1] += today_steps_participant[day]
            weekly_goals[week-1] += daily_goal_participant[day]
            
        elif day_of_week_participant[day] == 1:
            weekly_steps[week-1] += today_steps_participant[day]
            weekly_goals[week-1] += daily_goal_participant[day]
            
        elif day_of_week_participant[day] == 2:
            weekly_steps[week-1] += today_steps_participant[day]
            weekly_goals[week-1] += daily_goal_participant[day]
            
        elif day_of_week_participant[day] == 3:
            weekly_steps[week-1] += today_steps_participant[day]
            weekly_goals[week-1] += daily_goal_participant[day]
            
        elif day_of_week_participant[day] == 4:
            weekly_steps[week-1] += today_steps_participant[day]
            weekly_goals[week-1] += daily_goal_participant[day]
            
        elif day_of_week_participant[day] == 5:
            weekly_steps[week-1] += today_steps_participant[day]
            weekly_goals[week-1] += daily_goal_participant[day]
            
        elif day_of_week_participant[day] == 6:
            weekly_steps[week-1] += today_steps_participant[day]
            weekly_goals[week-1] += daily_goal_participant[day]
            
        count += 1   # If there is a day of the week present, add 1 count for the week
        weekly_counts[week-1] = count
        
        # This below is to calculate weekly average steps:
        weekly_average_steps[week-1] = weekly_steps[week-1]/weekly_counts[week-1] 
        
        # This below is to calculate weekly average goals:
        weekly_average_goals[week-1] = weekly_goals[week-1]/weekly_counts[week-1]
        
        
        # Condition to check to see if there is another day number that is bigger than the current day,
        # if there is, we continue onto the next day
        # if there isn't, we move onto the next week
        
        if day != len(day_of_week_participant) - 1:   # If day have not reached the last day of the range
            if day_of_week_participant[day+1] > day_of_week_participant[day]: # Then we compare if next day bigger than 
                continue                                               # current day, means the week havent ended
            else:                                           
                #weekly_counts[week-1] = count     # But if the next day is smaller or equal to current day, means we move
                week += 1                         # onto next week, and then we increase our week by 1 and reset the day 
                count = 0                         # counter to zero again.
        else:
            continue
        
    all_weekly_steps.append(weekly_steps)
    all_weekly_goals.append(weekly_goals)
    all_weekly_counts.append(weekly_counts)
    all_weekly_average_steps.append(weekly_average_steps)
    all_weekly_average_goals.append(weekly_average_goals)
    

#print(len(all_weekly_steps))
#print(len(all_weekly_goals))
#print(len(all_weekly_counts))
#print(len(all_weekly_average_steps))
#print(len(all_weekly_average_goals))
```

```python
####################################################################################################
########## Now need to find the frequency of motivational/feedback/timemsg for each week ###########
####################################################################################################

count = 0
all_feedback_counts = []
all_motivational_counts = []
all_timemsg_counts = []

for participant in df_participants:
    feedback_participant = new_dataset[new_dataset.ID_DIAMANTE == participant].feedback.reset_index(drop=True)
    motivational_participant = new_dataset[new_dataset.ID_DIAMANTE == participant].motivational.reset_index(drop=True)
    timemsg_participant = new_dataset[new_dataset.ID_DIAMANTE == participant].time_msg.reset_index(drop=True)
    
    weekly_feedback_count = []
    weekly_motivational_count = []
    weekly_timemsg_count = []
    
    index_start = 0
    end = 0

    for j in range(len(all_weekly_counts[count])):
        index_end = all_weekly_counts[count][j]
        end = end + index_end
        
        # For weekly feedback
        df_participant_feedback = dict(feedback_participant.loc[index_start:end-1].value_counts())
        weekly_feedback_count.append(df_participant_feedback)
        # For weekly motivational
        df_participant_motivational = dict(motivational_participant.loc[index_start:end-1].value_counts())
        weekly_motivational_count.append(df_participant_motivational)
        # For weekly time_msg
        df_participant_timemsg = dict(timemsg_participant.loc[index_start:end-1].value_counts())
        weekly_timemsg_count.append(df_participant_timemsg)

        # After that iteration: set index_start to previous index_end and index_end now to next index_end
        index_start += index_end
        
    count += 1
    
    all_feedback_counts.append(weekly_feedback_count)
    all_motivational_counts.append(weekly_motivational_count)
    all_timemsg_counts.append(weekly_timemsg_count)
    

# To create a single list for feedback
feedback_new_weekly = flatten(all_feedback_counts)
feedback_df_new = pd.DataFrame(feedback_new_weekly).fillna(0).astype(int)
#print(feedback_df)

# To create a single list for motivational
motivational_new_weekly = flatten(all_motivational_counts)
motivational_df_new = pd.DataFrame(motivational_new_weekly).fillna(0).astype(int)
#print(motivational_df)

# To create a single list for time_msg
timemsg_new_weekly = flatten(all_timemsg_counts)
timemsg_df = pd.DataFrame(timemsg_new_weekly).fillna(0).astype(int)
#print(time_msg_df)

#########################################################
#### Merge the above dataframes into 1 big dataframe ####
#########################################################
feedback_motivational_new = feedback_df_new.join(motivational_df_new)
print(feedback_motivational_new.head())
print(len(feedback_motivational_new))
```

```python
from sklearn.metrics import f1_score
from sklearn.pipeline import Pipeline
from yellowbrick.classifier import ClassificationReport
from sklearn.model_selection import cross_val_score

def cv_model_selection(X, y, model):
    
    # Instantiate the classification model 
    model.fit(X, y)
    
    expected  = y
    predicted = model.predict(X)   
    cv_scores = cross_val_score(model, X, y, cv = 10, scoring='neg_mean_squared_error') 
    
    # Calculate the F1 score
    F1 = f1_score(expected, predicted)
    
    model_results = pd.DataFrame({'Model Name': model, 'CV_mse': cv_scores.mean(), 'CV_std': cv_scores.std(),
                                 "F1_score":F1},index = [0])

    return model_results
```
